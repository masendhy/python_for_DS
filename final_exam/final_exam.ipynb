```python
def kalkulator(a,b):
    penjumlahan = a+b
    pengurangan = a-b
    perkalian = a*b
    return penjumlahan, pengurangan, perkalian

hasil = kalkulator(5,6)
print('Hasilnya adalah',hasil[1],hasil[0])
```

```python
def exam_python(is_achieve_order,year=2020):
    if year==2020:
        if is_achieve_order == True:
            result = "Food"
        else :
            result = 'Fashion'
    
    elif year==2021:
        if is_achieve_order == True:
            result = "Medical"
        else :
            result = "Electronic"
    else :
        result = "No-Category"
    return result
```

```python
exam_python(True, 2021)
```

```python
def extract_region_code(x):
    list_region_code = []
    for i in x:
        list_region_code.append(i.split()[0])
    return list_region_code
```
