```python
list_plat_mobil = ['B 1234 AB', 'B 6721 TY', 'B 1233 AY', 'B 6629 DD', 'B 1111 AM', 'B 6726 D', 'D 11223 KJ', 'AE 44677 GH', 'AE 67269 AA']

list_angka_plat = []

#ambil karakter angka dengan .isdigit()
for plat in list_plat_mobil:
    angka_plat = ""
    for karakter in plat:
        if karakter.isdigit():
            angka_plat += karakter
    list_angka_plat.append(angka_plat)


#ubah angka_plat menjadi int()
list_int_plat = []
for angka_plat in list_angka_plat:
  int_plat = int(angka_plat)
  list_int_plat.append(int_plat)

#hitung plat nomor genap dan ganjil
count_ganjil = 0
count_genap = 0

for i in list_int_plat:
  if i % 2 == 0:
    count_genap += 1
  else :
    count_ganjil += 1

print('Jumlah Plat nomor genap :', count_genap)
print('Jumlah Plat nomor ganjil :', count_ganjil)

    
```

```python
#. 3. Team Campaign PT ABC akan membuat promo bundling category product (berisi 3 category setiap bundling) untuk promo 17 Agustus. PT ABC memiliki 10 category product. Ketentuan Bundling:
#1. Tidak boleh hanya memiliki 1 jenis category dalam 1 bundling
#2. 1 Bundling boleh memiliki 2 category yang sama
#3. Urutan tidak menjadi faktor pembeda (c1 - c2 - c3 = c2 - c1 - c3).
#Tentukan ada berapa jumlah kemungkinan bundling yang bisa terbentuk?

list_id_category = ['c1', 'c2', 'c3', 'c4', 'c5', 'c6', 'c7' , 'c8', 'c9', 'c10']
list_building = []

for i in list_id_category:
  for y in list_id_category:
    for z in list_id_category:
      new_list = [i,y]
      if z not in new_list:
        new_list = [i,y,z]
        new_list.sort()
        if new_list not in list_building:
          list_building.append(new_list)


print(list_building)

print('Hasil kombinasi bundling ada : ',len(list_building) )

```

```python
# 4. Buatlah sebuah function string_processor dengan inputan 2 parameter, yaitu suatu string 1 kata dan nama proses yang akan dilakukan.

def string_processor(metode,string):
  string = string.lower()
  if metode =='first char':
    result = string[0]
  elif metode =='last char':
    result = string[-1]
  elif metode == 'num_chars':
    result = len(string)
  elif metode == 'num_vowels':
    vowels = 'aiueo'
    count = 0
    for char in string:
      if char in vowels:
        count += 1
    return count
  else :
    result = "Invalid process string"
  return result


```

```python
df_superstore = pd.read_excel('superstore.xls')
df_superstore.head()
```

```python
df_superstore['year'] = df_superstore['Order Date'].dt.year
```

```python
df_superstore = df_superstore[['year','Order ID', 'Order Date','Region','Sales','Profit']]
df_superstore
```

```python
def total_sales_profit(region, year = 2017):

  # filter data by year and region
  df_superstore_filtered = df_superstore[(df_superstore['Region']== region) & (df_superstore['year'] == year)]

  # calculate total sales and profit
  total_sales = round(df_superstore_filtered['Sales'].sum(),2)
  total_profit = round(df_superstore_filtered['Profit'].sum(),2)
    
  return (total_sales, total_profit)
```

```python
total_sales, total_profit = total_sales_profit(region='South', year = 2016)
print("Total sales di South Region pada tahun 2016: ", total_sales)
print("Total profit di South Region pada tahun 2016: ", total_profit)
```

```python
total_sales, total_profit = total_sales_profit(region='West')
print("Total sales di West Region pada tahun 2017:", total_sales)
print("Total profit di West Region pada tahun 2017:", total_profit)
```

```python
# 7. Menggunakan dataset superstore! Buatlah 3 kelompok 'segment' (high, medium, dan low) berdasarkan performance Sales yang digunakan sebagai acuan tim bisnis untuk mengelompokan customer superstore.

data_storesuper = pd.read_excel('superstore.xls')
data_storesuper.head()
```

```python
data_storesuper['year'] = data_storesuper['Order Date'].dt.year
data_storesuper
```

```python

df_group_storesuper = data_storesuper.groupby(['year'])['Sales'].describe(percentiles = [.25,.50,.75,.9,.95]).reset_index()
df_group_storesuper
```

```python
df_merge = pd.merge(data_storesuper,df_group_storesuper,on='year', how='inner')
df_merge
```

```python
df_merge.describe(percentiles = [.25,.50,.75,.9,.95]).reset_index()
```

```python
# create new column 'segment'
avg_sales = df_merge['Sales'].mean()
df_merge['segment'] = df_merge['Sales'].apply(lambda x: 'high' if x > 2*avg_sales else ('medium' if x > avg_sales else 'low'))
```

```python
df_merge[['year','Order ID', 'Order Date','Region','Sales','Profit','segment']]
```

```python
#Alasan pengelompokan ini didasarkan pada perbandingan setiap nilai Sales dengan nilai rata-rata. 
#Jika nilai Sales lebih besar dari 2 kali nilai rata-rata, maka dikategorikan sebagai kelompok 'high', 
#jika lebih besar dari nilai rata-rata tetapi tidak mencapai 2 kali nilai rata-rata, maka dikategorikan sebagai kelompok 'medium', 
#dan jika kurang dari atau sama dengan nilai rata-rata, maka dikategorikan sebagai kelompok 'low'. 
#Dengan pengelompokan ini, dapat memberikan gambaran yang lebih jelas tentang performa Sales tiap kelompoknya, 
#sehingga dapat membantu dalam menentukan strategi pemasaran dan penjualan pada masing-masing kelompok.
```

```python
#jumlah customer pada masing-masing segment di tahun 2017

df_merge_2017 = df_merge[df_merge['year'] == 2017]
customer_count = df_merge_2017.groupby('segment').size()

customer_count
```
